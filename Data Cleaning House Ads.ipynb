```python
import pandas as pd
import re
```

```python
df = pd.read_excel('C:/Users/sunma/Desktop/Intern/house ads/House Ads.xlsx')
```

```python
df.head()
```

```python
allowed_countries = ['Canada', 'United States']
df = df[df['Country'].isin(allowed_countries)]
```

```python
df.shape
```

```python
df.tail()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
aggregated = df.groupby([ 'Date', 'Creative', 'Country']).agg({'Total impressions': 'sum', 'Ad server clicks': 'sum'}).reset_index()
aggregated.tail(50)
```

```python
aggregated.shape
```

```python
aggregated.head(5)
```

```python
aggregated['Creative'] = aggregated['Creative'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))
```

```python
aggregated.head(5)
```

```python
aggregated.to_csv('aggregated_houseads.csv', index=False)
```
